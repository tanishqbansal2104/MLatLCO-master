```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt
import io
from PIL import Image
import requests

#Reading Image To Detect The Face
img = plt.imread("casual-eyeglasses-eyewear-2117283.jpg")
plt.imshow(img)
```

```python
#Converting Image From BGR to RGB
imgrgb=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(imgrgb)
```

```python
#Now Converting The Image From RGB To Fray
gray = cv.cvtColor(imgrgb,cv.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
detector = cv.CascadeClassifier("haar_frontal.xml")
faces = detector.detectMultiScale(gray,1.3,5)
x,y,w,h = faces[0]
cv.rectangle(img,(x,y),(x+w,y+h),[255,0,0],8)
plt.imshow(img)
```
