```python
from sklearn.datasets import load_breast_cancer
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
l=load_breast_cancer()
```

```python
# california_housing.get_data_home
```

```python
df =pd.DataFrame(data=load_breast_cancer().data,columns=load_breast_cancer().feature_names)
```

```python
df.columns
```

```python
df['label']=l.target
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr())
```

```python
# x=df[["mean fractal dimension","texture error","smoothness error","symmetry error","fractal dimension error"]].values
y=df["label"].values
x=df.iloc[:,:-1].values
```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=22)
```

```python
models=[DecisionTreeClassifier(),GaussianNB(),RandomForestClassifier(),SVC()]
def Model_fun(x_train,x_test,y_train,y_test):
    
    best_models=[]
    for i in models:
        i.fit(x_train,y_train)
        prediction=i.predict(x_test)
        accuracy=accuracy_score(prediction,y_test)*100
        print("Accuracy of" ,str(i).split("(")[0]," = ",accuracy)
        best_models.append(accuracy)
    return max(best_models)
```

```python
Model_fun(x_train,x_test,y_train,y_test)
```
