```python
import matplotlib.pyplot as plt
import requests
import json
```

```python
import requests
import json
temp = []
city = []
weather = []
k = 0
caps = ["Jhansi","Nagpur","Bilaspur","Vijayawada","Rentachintala","Surat","Vadodara","Lucknow","Jaipur","Kolkata"]
for i in caps:
    z = requests.get('http://api.openweathermap.org/data/2.5/weather?q='+i+'&APPID=9c28a85dca23d12779f571c379576d00')
    dic= json.loads(z.content.decode())
    print(type(z.content.decode()))
    if dic['cod'] == 200:
        k+=1
        print(k)
        print('City Name :',dic['name'])
        print('Temprature : ',dic['main']['temp'])
        print('Weather :',dic['weather'][0]['main'])
        temp.append(dic['main']['temp'])
        city.append(dic['name'])
        weather.append(dic['weather'][0]['main'])
    else:
        print(i,'Spelling is Incorrect')
```

```python
import pandas as pd
dt=pd.DataFrame({"City":city,"temp":temp,"weather":weather})
dt
```

```python
plt.figure(figsize=(10,10))
plt.plot(temp,city,color='green',linestyle='dashed',linewidth=3,marker='o',markerfacecolor='blue',markersize=10)
plt.xlabel('tempearture In Kelvin ')
plt.ylabel('State Capitals of India ')
plt.title('Temperature Varitation by State Capitals on IN')
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x=dt.City,y=dt.temp,hue=dt.weather)
plt.xticks(rotation=90)
plt.xlabel('City')
plt.ylabel('Temp')
plt.title('Temprature of Capitals')
# plt.legend(loc='lower right')
```

# point plot graph

```python
# f,ax1 = plt.subplot(figsize=(20,20))
plt.figure(figsize=(15,6))
sns.pointplot(x=dt.City,y=dt.temp,color='lime')
plt.xlabel('State',fontsize=15,color="red")
plt.ylabel('Temp',fontsize=15,color='blue')
plt.title("State Vs temperature ",fontsize=20,color='blue')
plt.xticks(rotation=90)
plt.grid()
```
